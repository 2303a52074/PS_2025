```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 30
std_dev = 7
N = 1000
data = np.random.normal(mean, std_dev, N)
plt.figure(figsize=(9, 5))
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Data Histogram')
x = np.linspace(min(data), max(data), 1000)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, color='red', linewidth=2, label='Normal Distribution PDF')

plt.title('Histogram and PDF of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, shapiro
mean = 30
std_dev = 7

p1 = 1 - norm.cdf(19, mean, std_dev)
p2 = norm.cdf(19, mean, std_dev)
p3 = norm.cdf(35, mean, std_dev) - norm.cdf(19, mean, std_dev)

x = np.linspace(10, 50, 1000)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'b', label='Normal Distribution PDF')

plt.fill_between(x, pdf, where=(x >= 19), color='pink', alpha=0.5, label='X ≥ 19')
plt.fill_between(x, pdf, where=(x <= 19), color='green', alpha=0.5, label='X ≤ 19')
plt.fill_between(x, pdf, where=(x >= 19) & (x <= 35), color='yellow', alpha=0.5, label='19 ≤ X ≤ 35')

plt.title('Normal Distribution with Probability Regions')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"P(X ≥ 19): {p1:.3f}")
print(f"P(X ≤ 19): {p2:.3f}")
print(f"P(19 ≤ X ≤ 35): {p3:.3f}")
```
