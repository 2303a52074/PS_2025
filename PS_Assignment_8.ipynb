```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

calorie_intake = np.array([2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000])
exercise_duration = np.array([60, 50, 30, 90, 45, 20, 75, 95, 15, 10])
bmi = np.array([25, 26, 28, 23, 27, 30, 24, 22, 31, 33])

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
axes[0].scatter(calorie_intake, bmi, color='b', alpha=0.7)
axes[0].set_xlabel('Daily Calorie Intake')
axes[0].set_ylabel('BMI')
axes[0].set_title('Daily Calorie Intake vs. BMI')
axes[1].scatter(exercise_duration, bmi, color='r', alpha=0.7)
axes[1].set_xlabel('Exercise Duration (minutes)')
axes[1].set_ylabel('BMI')
axes[1].set_title('Exercise Duration vs. BMI')
axes[2].scatter(calorie_intake, exercise_duration, color='g', alpha=0.7)
axes[2].set_xlabel('Daily Calorie Intake')
axes[2].set_ylabel('Exercise Duration (minutes)')
axes[2].set_title('Daily Calorie Intake vs. Exercise Duration')
plt.tight_layout()
plt.show()
```

```python
corr_calorie_bmi, _ = pearsonr(calorie_intake, bmi)
corr_exercise_bmi, _ = pearsonr(exercise_duration, bmi)
corr_calorie_exercise, _ = pearsonr(calorie_intake, exercise_duration)
print(f"Calorie Intake vs. BMI: {corr_calorie_bmi:.2f}")
print(f"Exercise Duration vs. BMI: {corr_exercise_bmi:.2f}")
print(f"Calorie Intake vs. Exercise Duration: {corr_calorie_exercise:.2f}")
```

```python
if corr_calorie_bmi > 0:
    print("Higher calorie intake is associated with a higher BMI.")
if corr_exercise_bmi < 0:
    print("More exercise duration is associated with a lower BMI.")
if corr_calorie_exercise < 0:
    print("Higher calorie intake is associated with lower exercise duration.")
```
