```python
import numpy as np
import matplotlib.pyplot as plt
N = 500
calls_per_day = np.random.randint(0, 8, size=N)
unique_calls, counts = np.unique(calls_per_day, return_counts=True)
pmf = counts / N
cdf = np.cumsum(pmf)
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].bar(unique_calls, pmf, color='green', edgecolor='black')
ax[0].set_title('PMF (Probability Mass Function)')
ax[0].set_xlabel('Number of Calls')
ax[0].set_ylabel('Probability')
ax[1].step(unique_calls, cdf, where='post', color='orange', linewidth=2)
ax[1].set_title('CDF (Cumulative Distribution Function)')
ax[1].set_xlabel('Number of Calls')
ax[1].set_ylabel('Cumulative Probability')
plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats
N = 100
p = 0.25
k_exactly_20 = 20
prob_exactly_20 = stats.binom.pmf(k_exactly_20, N, p)
k_at_least_30 = 30
prob_at_least_30 = 1 - stats.binom.cdf(k_at_least_30 - 1, N, p)
expected_value = N * p
print(f"i) Probability that exactly 20 patients need surgery: {prob_exactly_20:.4f}")
print(f"ii) Probability that at least 30 patients need surgery: {prob_at_least_30:.4f}")
print(f"iii) Average number of patients requiring surgery: {expected_value}")
```
